```python
def Rsquare(x,y,f):
    summean= sum((y-np.mean(y))**2)
    sumdev= sum((y-f(x))**2)
    return (1-(float(sumdev)/summean))
```

```python
# count to get hotspots based on 2016 data

#  a list num16 to save the total number of crime in every cell in 2016( loop for colunm45-55)

pre_data16=[]
for ii in range(0,12):
     num16=[]
     #for index, row in df_nbm.iterrows():
     for index, row in df_nbm.iterrows():
            
        
         num16.append(row[44+ii])
    

     # count how many grid in sum16 not equal to zero, just keep the ones which is not zero with their index in num16notzero
     n_notzero=0
     num16notzero=[]
     for i,val in enumerate(num16):
             if val != 0:
                  num16notzero.append([i,val])
                  n_notzero = n_notzero+1
        
        
     #print n_notzero, len(num16),len(num16notzero)

     # num16hots contains tuples of hotspots as [index,number]
     
     threshold=20
     while threshold > 3:#range to test threshold to decide which is better   
        num16hots=[]
        for i,val in enumerate(num16notzero):
            if val[1] > threshold:
        
                num16hots.append(val)
       
        x=[] # x for linear regression
        lg=[] # contains the polyfit for each grid
        for i in range(0,44+ii):
           x.append(i)
        for index, row in df_nbm.iterrows():
           p=np.polyfit(x,row[0:(44+ii)],1)
           lg.append(p)
    
        
        phots=[] # contains the predict result for each grid
        for i,val in enumerate(lg):
           f = np.poly1d(val) 
                 
     
            
           phots.append([f(44+ii),Rsquare(x,df_nbm.as_matrix()[i][0:(44+ii)],f)])   

        phots_notzero=[] #contains the grids(index, pedict result ) which predict result is not zero
        evaR2all=0
        for i,val in enumerate(phots):
            evaR2all=evaR2all+val[1]
            if val[0] != 0:
              phots_notzero.append([i,val[0],val[1]])
        
     
     
        phots_f=[]
        evaR2=0
        for i,val in enumerate(phots_notzero):
            if val[1] > threshold:
                evaR2=evaR2+val[2]
                phots_f.append(val)              
        threshold=threshold-1
        if ((len(phots_f) > 20) & (len(phots_f) <= 58) & (len(num16hots) <= 58) & (len(num16hots) > 20)) :
            break
       
        

     
     validation=0
     for i,v in enumerate(num16hots):
            for inx,val in enumerate(phots_f):
                if v[0] == val[0]:
                    validation=validation +1
     
     
                        
     
     evaR2all= float(evaR2all)/len(phots)                                    
     evaR2= float(evaR2)/len(phots_f)                    
     pre_data16.append([len(num16hots),len(phots_f),validation, float(validation)/len(num16hots),evaR2all,evaR2,threshold+1])       
```
